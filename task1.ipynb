## Objective
To load the Iris dataset, explore its structure, and create meaningful visualizations that reveal patterns across the three flower species: Setosa, Versicolor, and Virginica.

```python
# Step 2: Import all required libraries
import pandas as pd         # for working with data tables
import numpy as np          # for numbers and math
import matplotlib.pyplot as plt # for creating charts
import seaborn as sns       # for prettier charts

print("Libraries imported successfully!")
```

```python
# Step 3: Load the Iris dataset directly from seaborn (no download needed)
df = sns.load_dataset('iris')

# Show the first 5 rows
print("First 5 rows of the dataset:")
df.head()
```

```python
# Explore the dataset structure
print("Shape (rows, columns):", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nData types and info:")
df.info()
```

```python
# Statistical summary
print("Statistical Summary:")
display(df.describe())

print("\nFlowers per species:")
print(df['species'].value_counts())
```

```python
# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    palette='Set1'
)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Histograms for all 4 measurements
df.drop('species', axis=1).hist(
    bins=20,
    figsize=(10, 6),
    color='steelblue',
    edgecolor='black'
)
plt.suptitle('Distribution of All Features', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Box plots for each feature grouped by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    sns.boxplot(data=df, x='species', y=feature, palette='pastel', ax=ax)
    ax.set_title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```

## Summary of Findings

- The dataset has **150 rows and 5 columns** (4 features + 1 species label).
- There are **3 species**: Setosa, Versicolor, and Virginica — 50 flowers each.
- The **scatter plot** shows Setosa is clearly separated from the other two species based on sepal measurements.
- **Histograms** reveal that petal length and petal width have two distinct groups, suggesting strong separability.
- **Box plots** confirm that Setosa has significantly smaller petals than Versicolor and Virginica.
- No major outliers were detected across the four features.
